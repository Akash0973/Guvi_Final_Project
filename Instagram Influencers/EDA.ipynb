```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

data=pd.read_csv('C:\Guvi_final_project\Instagram Influencers/Influencer.csv')
```

```python
def cleaner(value):
    suf=value[-1]
    num=value[0:-1]
    if suf=='k':
        return (1000*float(num))
    elif suf=='m':
        return (1000000*float(num))
    elif suf=='b':
        return (1000000000*float(num))
    else:
        return float(value)

def name_cleaner(name):
    name=name.replace('\n','')
    return name
```

```python
data['Followers']=data['Followers'].apply(cleaner)
data['Avg. Likes']=data['Avg. Likes'].apply(cleaner)
data['Posts']=data['Posts'].apply(cleaner)
data['New Post Avg. Likes']=data['New Post Avg. Likes'].apply(cleaner)
data['Total Likes']=data['Total Likes'].apply(cleaner)
data['Channel Info']=data['Channel Info'].apply(name_cleaner)
```
