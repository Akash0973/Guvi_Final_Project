```python
import pickle 
import matplotlib.pyplot as plt

with open('Model3_test.pkl','rb') as file:
            test_acc0=pickle.load(file)
with open('Model3_train.pkl','rb') as file:
            train_acc0=pickle.load(file)
with open('Model4_test.pkl','rb') as file:
            test_acc1=pickle.load(file)
with open('Model4_train.pkl','rb') as file:
            train_acc1=pickle.load(file)
with open('Model5_test.pkl','rb') as file:
            test_acc2=pickle.load(file)
with open('Model5_train.pkl','rb') as file:
            train_acc2=pickle.load(file)
```

```python
plt.figure(figsize=(8, 6))
plt.title("Without Dropout")
plt.xlabel("Epochs")
plt.ylabel("Accuracy")

plt.plot([i for i in range(1,31)], test_acc0, label='Test accuracy', color='blue', marker='o')
plt.plot([i for i in range(1,31)], train_acc0, label='Train accuracy', color='green', marker='s')

plt.legend()
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.title("With Dropout 0.2")
plt.xlabel("Epochs")
plt.ylabel("Accuracy")

plt.plot([i for i in range(1,31)], test_acc1, label='Test accuracy', color='blue', marker='o')
plt.plot([i for i in range(1,31)], train_acc1, label='Train accuracy', color='green', marker='s')

plt.legend()
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.title("With Dropout 0.4")
plt.xlabel("Epochs")
plt.ylabel("Accuracy")

plt.plot([i for i in range(1,31)], test_acc2, label='Test accuracy', color='blue', marker='o')
plt.plot([i for i in range(1,31)], train_acc2, label='Train accuracy', color='green', marker='s')

plt.legend()
plt.show()
```
